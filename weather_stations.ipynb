```python
import pandas as pd
```

```python
file_path = "Resources/master_location_identifier.csv"
```

```python
master_loc_id_df = pd.read_csv(file_path, encoding = "ISO-8859-1")
```

```python
master_loc_id_df.head()
```

```python
station_id_df = master_loc_id_df[['country2', 'country', 'wban', 'lat', 'lon']].copy()
```

```python
station_id_df = station_id_df.rename(columns = {'country2':'Country ID', 'country': 'Country', 
                                                'wban': 'Station ID', 'lat': 'Lat', 'lon': 'Lng'})
station_id_df = station_id_df.set_index('Country ID')
```

```python
station_id_df
```

```python
only_us = station_id_df.loc[station_id_df["Country"] == "United States", :]
```

```python
us_station_id = station_id_df[station_id_df.index == 'US']
us_station_id = us_station_id.dropna()
```

```python
us_station_id["Station ID"] = us_station_id["Station ID"].astype(int) 
```

```python
us_station_id
```

```python
us_station_id = us_station_id.set_index('Station ID')
```

```python
us_station_id = us_station_id.drop(columns = ['Country'])
```

```python
us_station_id
```

```python
us_station_id.to_csv('Resources/us_station_id.csv', index = False)
```
